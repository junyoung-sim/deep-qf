```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_pacf
from arch import arch_model
from download import download

apikey = open("apikey", "r").readline()
```

```python
QQQ = download("QQQ", apikey).rename(columns={"adjClose": "QQQ"}) # NASDAQ Index (Growth)
DIA = download("DIA", apikey).rename(columns={"adjClose": "DIA"}) # Dow Index (Value)

dat = QQQ.merge(DIA, on="date")
print(dat)
```

```python
t0 = np.array(dat["QQQ"][0:-1])
t1 = np.array(dat["QQQ"][1:])
QQQ = (t1 - t0) * 100 / t0 # QQQ Daily Returns (%)

t0 = np.array(dat["DIA"][0:-1])
t1 = np.array(dat["DIA"][1:])
DIA = (t1 - t0) * 100 / t0 # DIA Daily Returns (%)

plt.plot(QQQ, label="QQQ", color="steelblue")
plt.plot(DIA, label="DIA", color="cadetblue")
plt.legend()
plt.show()
```

```python
res = QQQ - DIA # return residual
plt.plot(res, label="QQQ - DIA", color="steelblue")
plt.legend()
plt.show()
```

```python
plot_pacf(res**2) # estimate correlation between lags within return residual
plt.show()
```

```python
lags = 2
model = arch_model(res, p=lags, q=lags)
```

```python
model.fit().summary()
```

```python
# rolling forecast (recent 1000 days)
pred = []
start = res.shape[0] - 1000
for i in range(start, res.shape[0]):
    test = res[:i]
    model = arch_model(test, p=lags, q=lags)
    yhat = np.sqrt(model.fit(disp="off").forecast(horizon=1).variance.values[-1,:][0])
    pred.append(yhat)
```

```python
plt.figure(figsize=(15,5))
plt.plot(res[start:], label="QQQ - DIA")
plt.plot(pred, color="red", label="Predicted Return Residual Volatility")
plt.legend()
plt.show()
```

```python
# real-time prediction
model = arch_model(res, p=lags, q=lags)
yhat = np.sqrt(model.fit(disp="off").forecast(horizon=5).variance.values[-1,:])
plt.plot(np.insert(yhat, 0, pred[-1]))
plt.show()
```
